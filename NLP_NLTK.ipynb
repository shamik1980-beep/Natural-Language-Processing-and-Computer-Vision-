```python
import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk import pos_tag

# Sample text
text = "Self-driving cars use NLP and computer vision. Pedestrians are detected at close distances. Safety is critical."

print("ORIGINAL TEXT:")
print(text)


# 1. Sentence Tokenization

sentences = sent_tokenize(text)
print("\nSENTENCES:")
for s in sentences:
    print("-", s)


# 2. Word Tokenization

tokens = word_tokenize(text)
print("\nTOKENS:")
print(tokens)


# 3. Cleaning

clean_tokens = [w.lower() for w in tokens if w.isalpha()]
print("\nCLEAN TOKENS:")
print(clean_tokens)


# 4. Stopword Removal

stop_words = set(stopwords.words("english"))
filtered = [w for w in clean_tokens if w not in stop_words]
print("\nAFTER STOPWORD REMOVAL:")
print(filtered)


# 5. Lemmatization

lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w) for w in filtered]
print("\nLEMMATIZED WORDS:")
print(lemmatized)



# 6. Sentiment Analysis

sia = SentimentIntensityAnalyzer()
sentiment = sia.polarity_scores(text)
print("\nSENTIMENT:")
print(sentiment)
```
